```python
import pandas as pd
StudentsPerformance = pd.read_csv('StudentsPerformance.csv')
```

```python
import matplotlib.pyplot as plt
reading_score = StudentsPerformance ['reading_score']
math_score = StudentsPerformance['math_score']
```

```python
plt.scatter(reading_score, math_score)
plt.xlabel('reading_score')
plt.ylabel('math_score')
plt.title('correlation between Math Score and Reading Score')
```

```python
import scipy.stats
scipy.stats.pearsonr(reading_score, math_score)
```
